# ***Import***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import files
data = files.upload()
```

```python
import io
data = pd.read_csv(io.BytesIO(data['ionosphere_data.csv']))
```

# ***Load data***

```python
data
```

```python
targetEncoded = []
for i in range(len(data["label"].values)):
    if data["label"].values[i] == "g":
        value=1
    else:
        value=0
    targetEncoded.append(value)
data["target_encoded"] = targetEncoded
data.drop(["label"], axis=1,inplace=True)
```

```python
y = data["target_encoded"].values
X = data.copy(deep=True)
X.drop(["target_encoded"], axis=1, inplace=True)
X_train = X.values[0:210]
X_test = X.values[210:]
y_train = y[0:210]
y_test = y[210:]
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers

network = models.Sequential()

network.add(layers.Dense(16, activation = 'relu', input_shape = (34,)))
network.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
network.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
network.fit(X_train, y_train, epochs=100, batch_size=128)
```


# ***Retraining the model with epochs=500***

```python
from tensorflow.keras import models
from tensorflow.keras import layers

network = models.Sequential()

network.add(layers.Dense(16, activation = 'relu', input_shape = (34,)))
network.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
network.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
network.fit(X_train, y_train, epochs=500, batch_size=128)
```

```python
test_loss, test_accuracy = network.evaluate(X_test,y_test)
print("test accuracy is :", test_accuracy)
print("test Loss is :", test_loss)
```
